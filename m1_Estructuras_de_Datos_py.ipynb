```python
doble = lambda x: x * 2
print(doble(8)) # 16
# Útil como argumento en funciones de orden superior
nums = [5, 2, 8, 1, 9, 3]
ordenados = sorted(nums, key=lambda x: -x) # orden descendente
# [9, 8, 5, 3, 2, 1]
# ── map y filter ────────────────────────────────────────────────────
cuadrados = list(map(lambda x: x**2, nums))
mayores = list(filter(lambda x: x > 4, nums))
```

```python
#Ejercicio 1: Crea una lista con los múltiplos de 7 entre 1 y 100 usando comprehension.
multiplos_7 = [x for x in range(1, 101) if x % 7 == 0]
print(multiplos_7)
```

```python
#Ejercicio 2: Dado el diccionario datos = {'a':10, 'b':25, 'c':7, 'd':18}
#a) Crea una lista con las claves cuyo valor sea mayor a 12.
#b) Crea un nuevo diccionario con esas claves y sus valores al cuadrado.
datos = {'a':10, 'b':25, 'c':7, 'd':18}

# a) Claves con valor mayor a 12
claves_mayores = [k for k, v in datos.items() if v > 12]
print(claves_mayores)

# b) Nuevo diccionario con valores al cuadrado
dict_cuadrado = {k: v**2 for k, v in datos.items() if v > 12}
print(dict_cuadrado)
```

```python
#Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con:
#longitud, suma, promedio, mínimo y máximo de la lista.
def resumen(lista):
    return {
        'longitud': len(lista),
        'suma': sum(lista),
        'promedio': sum(lista) / len(lista),
        'minimo': min(lista),
        'maximo': max(lista)
    }

print(resumen([3, 1, 7, 2, 5]))
```

```python
#Ejercicio 4: Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5].
def potencia(x, exp=3):
    return x ** exp

lista = [1, 2, 3, 4, 5]
resultado = list(map(lambda x: potencia(x), lista))
print(resultado)
```
