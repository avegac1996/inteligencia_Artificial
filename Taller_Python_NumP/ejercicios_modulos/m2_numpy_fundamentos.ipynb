```python
import numpy as np
```

```python
#Ejercicio 1: Crea los siguientes arrays y verifica sus atributos (shape, dtype, ndim):
#a) Vector con enteros del 1 al 20.
#b) Matriz de ceros de 5 filas y 3 columnas.
#c) 10 puntos igualmente espaciados entre -1 y 1 con linspace.
# a) Vector con enteros del 1 al 20
vector = np.arange(1, 21)
print('Vector:', vector)
print('shape:', vector.shape, 'dtype:', vector.dtype, 'ndim:', vector.ndim)

# b) Matriz de ceros 5x3
matriz_ceros = np.zeros((5, 3))
print('Ceros shape:', matriz_ceros.shape, 'dtype:', matriz_ceros.dtype)

# c) 10 puntos entre -1 y 1
puntos = np.linspace(-1, 1, 10)
print('Linspace:', puntos)
print('shape:', puntos.shape, 'dtype:', puntos.dtype)
```

```python
#Ejercicio 2: Con np.random.seed(99) crea una matriz 4×4 de enteros entre 0 y 50.
#Imprime: shape, dtype, tamaño en bytes (nbytes) y el elemento [2,3].
np.random.seed(99)
matriz = np.random.randint(0, 51, size=(4, 4))
print(matriz)
print('shape:', matriz.shape)
print('dtype:', matriz.dtype)
print('nbytes:', matriz.nbytes)
print('elemento [2,3]:', matriz[2, 3])
```

```python
#Ejercicio 3: ¿Cuál es la diferencia de dtype entre np.array([1,2,3]) y
#np.array([1.0,2.0,3.0])? Verifica con código.
enteros = np.array([1, 2, 3])
flotantes = np.array([1.0, 2.0, 3.0])

print('Enteros dtype:', enteros.dtype)
print('Flotantes dtype:', flotantes.dtype)
```

```python
#Ejercicio 4: Crea una matriz identidad 5×5 y verifica que la suma de su diagonal
#es igual a 5 (usa np.trace())
identidad = np.eye(5)
print(identidad)
print('Suma diagonal (trace):', np.trace(identidad))
```
